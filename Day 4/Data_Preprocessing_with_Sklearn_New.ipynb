```python
import pandas as pd
import numpy as np
df=pd.read_csv('Data.csv')
print(df)

print(df.isnull().sum())

# #filling
df['Age'].fillna(df['Age'].mean(),inplace=True)# fillna: a pandas dataframe used to fill missing nan values

df['Salary'].fillna(df['Salary'].median(),inplace=True)
# he inplace=True parameter ensures that the operation
#modifies the DataFrame df directly, without needing to reassign it to df.
print(df)
```

```python
df['Age']=pd.to_numeric(df['Age'])
#pd.to_numeric: used to convert a column to numeric value (either int or float)
df['Salary']=pd.to_numeric(df['Salary'])
print(df.dtypes)
```

```python
# Step 4: Normalize /Scale numeric values
#normalizing age and salary columns using min-max scaling.
'''Normalizing: it is a process of scaling the values between the range of 0 and 1,
Many machine learning algorithms, such as gradient-based optimization algorithms (like gradient descent), are
sensitive to the scale of features. Features with larger scales
can dominate the learning process, leading to slower convergence or even failure to converge.'''

'''
Normalization in machine learning is the process of translating data into the range [0, 1]
 (or any other range) or simply transforming data onto the unit sphere. '''

from sklearn.preprocessing import MinMaxScaler
#MinMaxScaler is a class which is imported from sklearn.preprocessing module

scaler = MinMaxScaler() #creates an instance of MinMaxScaler, which will be used to scale the numeric columns.

# normalize age and salary
df[['Age','Salary']]=scaler.fit_transform(df[['Age','Salary']])

print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler

data={
    'Name':['John','Jane','Doe','Alice','Bob'],
    'Age':[28,35,None,30,25],
    'Salary':[50000,None,40000,70000,45000],
    'Department':['IT','HR','Marketing','IT','HR']
}

df=pd.DataFrame(data)

#Filling missing age with mean age
df['Age'].fillna(df['Age'].mean(),inplace=True)

#filling missing salary with median salary
df['Salary'].fillna(df['Salary'].median(),inplace=True)

#Convert age and salary to numberic types if they are not already
df['Age']=pd.to_numeric(df['Age'])
df['Salary']=pd.to_numeric(df['Salary'])


#Normalize Age and salary
scaler=MinMaxScaler()
df[['Age','Salary']]=scaler.fit_transform(df[['Age','Salary']])

#Convert the department column into dummy variable
df=pd.get_dummies(df,columns=['Department'],prefix='Dept')


print(df)
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)

# train the model

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

classifier_knn = KNeighborsClassifier(n_neighbors=3)
classifier_knn.fit(x_train,y_train)
y_pred=classifier_knn.predict(x_test)
#finding accuracy by comparing actual response values (y_test) with predicted value (y_pred)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

#Providing sample data and the model will make prediction out of that data
sample=[[5,5,3,2],[2,4,3,5]]
preds=classifier_knn.predict(sample)

pred_species=[iris.target_names[p] for p in preds]
print("Predictions:",pred_species)
```
