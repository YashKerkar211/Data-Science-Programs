```python
#importing dependencies
import numpy as np
import pandas as pd # We are creating dataframes, dataframes are structured tables
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("creditcard.csv")
```

```python
# Distribution of legit and fraud transaction
df['Class'].value_counts()
```

```python
# So it means 284315 are legit transaction and 492 transaction are fraud, the data is very imbalance here.
#separating data for analysis

legit=df[df.Class==0] # all legit rows will be stored in 'legit' variable
fraud=df[df.Class==1] # all fraud rows will be stored in 'fraud' variable
```

```python
print(legit.shape)
print(fraud.shape)  
```

```python
# statistical measures of data
legit.Amount.describe()
# count describing number of data points we have (rows)
```

```python
# 25% (percentile) of transactions are less than this 5.650000$
# 50% of transactions are less than this 22.000000$
# 75% of transactions are less than this 77.050000$
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transaction
df.groupby('Class').mean()
```

```python
# Handling imbalance data using Undersampling (randomsampling)
#Build a sample dataset containing similar distribution of  normal and fraud  tranaction

legit_sample=legit.sample(n=492) # it will select random 492 legit datapoints
#The sample() method returns a list with a specified number of randomly selected items from a sequence.
#This method does not change the original sequence.
```

```python
 # Concatenating both 492 legit and fraud transactions
new_dataset=pd.concat([legit_sample,fraud],axis=0) #axis=0 means all 492 dataframe values will be added 1 by 1 below the the legit sample
#axis=1 values will be added column wise.
```

```python
new_dataset.head()  # these are random picked values
```

```python
new_dataset.tail() # these are the fraud transaction which we are having from starting, these are added below legit transaction
```

```python
new_dataset['Class'].value_counts()
# Now we have uniformly distribute data
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting into data into features & target
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X)
```

```python
print(Y)
```

```python
# Splitting features and target in training and testing 
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2,stratify=Y,random_state=2) # 3 means data will be splitted in different way, when running
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
# Training model using Logistic regression: because for binary classification
from sklearn.linear_model import LogisticRegression
model=LogisticRegression() # loading 1 instance of logistic regression
# Training logistic regression model with training data
# for training we use fit() method
model.fit(X_train,Y_train) # X_train contains all the features data, Y_train containing all the classes corresponding to the feature data
```

```python
# Model Evaluation
# first find accuracy score on training data
from sklearn.metrics import accuracy_score
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train) # Comparing values predicted by model on X_train with the actual values in Y_train
```

```python
print("Accuracy of Logistic Regression on training data:\n",training_data_accuracy)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("creditcard.csv")

x = df['Time']
y = df['Amount']

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.5)
plt.title('Scatter Plot of Transaction Amount vs. Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Amount ($)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('creditcard.csv')  
class_counts = df['Class'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, alpha=0.75)
plt.title('Bar Plot of Transaction Class')
plt.xlabel('Class')
plt.ylabel('Number of Transactions')
plt.xticks(ticks=[0, 1], labels=['Non-Fraudulent', 'Fraudulent'])
plt.grid(axis='y')
plt.show()
```
